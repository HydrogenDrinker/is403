```python
import pandas as pd
import os
import sys
import numpy as np
import ast

def show_training_log(file_path, save_txt_path=None):
    """
    H√†m ƒë·ªçc file history (d·∫°ng dictionary ch·ª©a NaN v√† Infinity)
    """
    
    if not os.path.exists(file_path):
        print(f"‚ùå L·ªói: Kh√¥ng t√¨m th·∫•y file t·∫°i '{file_path}'")
        return

    try:
        with open(file_path, 'r') as f:
            content = f.read().strip()
            
        # --- PH·∫¶N S·ª¨A L·ªñI QUAN TR·ªåNG ---
        try:
            # ƒê·ªãnh nghƒ©a t·ª´ ƒëi·ªÉn ƒë·ªÉ Python hi·ªÉu c√°c t·ª´ l·∫° trong file
            context = {
                'NaN': np.nan, 
                'nan': np.nan,
                'Infinity': np.inf,  # <--- Th√™m d√≤ng n√†y ƒë·ªÉ hi·ªÉu V√¥ c·ª±c
                'inf': np.inf,
                '-Infinity': -np.inf
            }
            
            # D√πng eval v·ªõi context ƒë√£ m·ªü r·ªông
            # Thay ƒë·ªïi __builtins__: None th√†nh {} ƒë·ªÉ tr√°nh l·ªói 'NoneType is not subscriptable'
            data_dict = eval(content, {}, context)
            
            df = pd.DataFrame(data_dict)
            
        except Exception as e:
            print(f"‚ö†Ô∏è L·ªói parse Dictionary: {e}")
            print("üëâ ƒêang th·ª≠ ƒë·ªçc nh∆∞ CSV th√¥ng th∆∞·ªùng...")
            df = pd.read_csv(file_path)

        total_epochs = len(df)
        f_out = open(save_txt_path, 'w') if save_txt_path else None
        
        def log(text):
            print(text)
            if f_out: f_out.write(text + "\n")

        log("=" * 60)
        log(f"TRAINING LOG RECONSTRUCTION")
        log(f"Source: {os.path.basename(file_path)}")
        log(f"Total Epochs: {total_epochs}")
        log("=" * 60)
        log("") 

        if total_epochs == 0:
            print("‚ùå File r·ªóng.")
            return

        for i, row in df.iterrows():
            epoch = i + 1
            
            loss = row.get('loss', float('nan'))
            rse = row.get('rse', float('nan'))
            corr = row.get('corr', float('nan'))
            
            val_loss = row.get('val_loss', float('nan'))
            val_rse = row.get('val_rse', float('nan'))
            val_corr = row.get('val_corr', float('nan'))

            log(f"Epoch {epoch}/{total_epochs}")
            
            # Format chu·ªói, x·ª≠ l√Ω in ra 'inf' n·∫øu l√† v√¥ c·ª±c
            log_line = (f"val_loss={val_loss:.6f}, val_rse={val_rse}, val_corr={val_corr}, "
                        f"loss={loss:.6f}, rse={rse:.6f}, corr={corr:.6f}")
            
            log(log_line)
            log("-" * 60) 

        if f_out:
            print(f"\n‚úÖ ƒê√£ l∆∞u log ra file: {save_txt_path}")
            f_out.close()
            
    except Exception as e:
        print(f"‚ùå C√≥ l·ªói x·∫£y ra: {e}")
```
