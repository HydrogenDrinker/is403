```python
import pandas as pd

# Load hourly data
df = pd.read_csv("../data/traffic.txt", delimiter=",", header=None)

# Create hourly timestamps from 2015-01-01
start = pd.Timestamp("2015-01-01 00:00")
time_index = pd.date_range(start=start, periods=len(df), freq="H")
df.index = time_index
```

```python
hourly_mean = df.mean(axis=1)
```

```python
# Create features for day of week and hour
day = hourly_mean.index.dayofweek   # 0 = Monday, 6 = Sunday
hour = hourly_mean.index.hour

# Group by (day, hour) and compute mean
weekly_pattern = hourly_mean.groupby([day, hour]).mean()
```

```python
# Flatten into a continuous 7-day pattern (168 hours)
import numpy as np

weekly_pattern_values = weekly_pattern.values
```

```python
two_week_pattern = np.tile(weekly_pattern_values, 2)  # repeat twice
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(16,10))
plt.plot(two_week_pattern, color='tab:blue')

plt.title("Average 14-Day Occupancy Pattern (Aggregated Across All Sensors)")
plt.xlabel("Hour Index (14 days total)")
plt.ylabel("Occupancy Rate")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# --- Load and index data ---
df = pd.read_csv("../data/traffic.txt", delimiter=",", header=None)
start = pd.Timestamp("2015-01-01 00:00")
df.index = pd.date_range(start=start, periods=len(df), freq="H")

# --- Average across all sensors ---
hourly_mean = df.mean(axis=1)

# --- Select first 14 days (14 * 24 = 336 hours) ---
first_14_days = hourly_mean.iloc[:14*24]

# --- Build x-axis labels (Monâ€“Sun repeated twice) ---
days = ['Mon','Tue','Wed','Thu','Fri','Sat','Sun'] * 2
x_ticks = range(0, len(first_14_days), 24)

# --- Plot ---
plt.figure(figsize=(8,5))
plt.plot(first_14_days, 'b')
plt.xticks(x_ticks, days)
plt.ylabel('Occupancy rate (%)')
plt.title('Occupancy Rate â€“ First 14 Days of Dataset')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
